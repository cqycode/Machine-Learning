### 线性可分支持向量机
观察C取值对决策边界的影响

数据：（x1,x2）-> color

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data=sio.loadmat('./data/ex6data1.mat')
data.keys()
```

```python
X,y=data['X'],data['y']
X.shape,y.shape
```

### 损失函数

$$
J(θ)=C\sum_{i=1}^{m}[y^{(i)}cost_1(θ^⊤x)+(1−y^{(i)})cost_0(θ^⊤x)]+ \frac{1}{2}\sum_{j=1}^{n}θ_j^2
$$

C:误差项惩罚系统

C越大，容错越低，易过拟合，低偏差，高方差

```python
def plot_data():
    plt.scatter(X[:,0],X[:,1],c=y.flatten(),cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('y1')
```

```python
plot_data()
```

```python
from sklearn.svm import SVC
```

```python
svc1=SVC(C=1,kernel='linear')
svc1.fit(X,y.flatten())
```

```python
svc1.predict(X)
```

```python
svc1.score(X,y.flatten())
```

```python
def plot_boundary(model):
    x_min,x_max=-0.5,4.5
    y_min,y_max=1.3,5
    xx,yy=np.meshgrid(np.linspace(x_min,x_max,500),np.linspace(y_min,y_max,500))
    z=model.predict(np.c_[xx.flatten(),yy.flatten()])#np.c_[a,b]表示按行拼接 r_[a,b]按列拼接

    zz=z.reshape(xx.shape)
    plt.contour(xx,yy,zz)#等高线
```

```python
plot_boundary(svc1)
plot_data()
```

```python
svc100=SVC(C=100,kernel='linear')
svc100.fit(X,y.flatten()) 
```

```python
svc100.predict(X)
svc100.score(X,y.flatten())
```

```python
plot_boundary(svc100)
plot_data()
```
