### 神经网络手写数字识别
数据集：ex3data1.mat  
参数：ex3weights.mat

```python
import numpy as np
import scipy.io as sio
```

```python
data=sio.loadmat('ex3data1.mat')
raw_X=data['X']
raw_y=data['y']
```

```python
X=np.insert(raw_X,0,values=1,axis=1)#添加偏置项
X.shape
```

```python
y=raw_y.flatten()
y.shape
```

```python
theta=sio.loadmat('ex3weights.mat')
theta.keys()
```

```python
theta1=theta['Theta1']
theta2=theta['Theta2'] 
theta1.shape,theta2.shape
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

**偏置（Bias）**，用来调整神经元对输入的敏感程度。偏置项会加在输入值上，并且对**每个神经元**都有一个独立的偏置值。它的作用是在激活函数中引入一定的灵活性，使得神经网络能够学习到更复杂的函数关系。偏置可以理解为一种偏移量，它可以使神经元在输入空间中更灵活地调整位置，从而更好地拟合数据。

```python
a1=X
```

```python
z2=X@theta1.T
a2=sigmoid(z2)
a2.shape
```

```python
a2=np.insert(a2,0,values=1,axis=1)
a2.shape
```

```python
z3=a2@theta2.T
a3=sigmoid(z3)
a3.shape
```

```python
y_pred=np.argmax(a3,axis=1)
y_pred=y_pred+1
y_pred
```

```python
acc=np.mean(y_pred==y)
acc
```
